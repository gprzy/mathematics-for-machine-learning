## Funções utilizadas em Machine Learning

```python
import numpy as np
from math import e
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-5, 5, 100)
```

**Sigmóide ou logística**
$$y = \frac{1}{1 + e^{-x}}$$

```python
sigmoid = lambda x: 1 / (1 + e**(-x))
Y = [sigmoid(x) for x in X]

plt.title('Sigmóide ou Logística');
plt.plot(X, Y);
plt.xlabel('$X$');
plt.ylabel('$Y$');
```

**Tangente hiperbólica**
$$y = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python
tg_hiper = lambda x: (e**x - e**-x) / (e**x + e**-x)
Y = [tg_hiper(x) for x in X]

plt.title('Tangente hiperbólica');
plt.plot(X, Y);
plt.xlabel('$X$');
plt.ylabel('$Y$');
```

**Linear retificada (ReLU)**
$$y = \max{\{0,x\}}$$

```python
relu = lambda x: max(0, x)
Y = [relu(x) for x in X]

plt.title('ReLU');
plt.plot(X, Y);
plt.xlabel('$X$');
plt.ylabel('$Y$');
```

**Leaky ReLU**
$$y = \max{\{ax,x\}}$$

```python
leaky_relu = lambda x, a=2: max(a*x, x)
Y = [leaky_relu(x) for x in X]

plt.title('Leaky ReLU');
plt.plot(X, Y);
plt.xlabel('$X$');
plt.ylabel('$Y$');
```
